```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create

```

```python
df = pd.read_csv('movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there is any missing data
df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Created correct Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')
plt.show()

```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'black'},line_kws={'color':'red'})
```

```python
# Let's start looking at correlation
```

```python
df.corr(method='pearson',numeric_only=True) #pearson, kendall, spearman
```

```python
# High correlatiion between budget and gross
```

```python
corrlation_matrix = df.corr(method='pearson',numeric_only=True)

sns.heatmap(corrlation_matrix, annot=True)


plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looks at Company

df.head()
```

```python
df_numerized = df.copy() 

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
corrlation_matrix = df_numerized.corr(method='pearson',numeric_only=True)

sns.heatmap(corrlation_matrix, annot=True)


plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)> 0.5]

high_corr
```
